```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
data = pd.read_csv('/Users/adamafaal-refaee/Desktop/Desktop-Win/codeing/projects/machine_learning/handmarks/data/not_high_5.csv')
col1 = data.iloc[:, 0]
col2 = data.iloc[:, 1]
col3 = data.iloc[:, 2]
col4 = data.iloc[:, 3]
```

```python
differences_x = []
differences_y = []

filename = "/Users/adamafaal-refaee/Desktop/Desktop-Win/codeing/projects/machine_learning/handmarks/data/not_average.csv"

columns = ["Cycle", "Plane"] + [f"Landmark #{i+1} - #{i+2}" for i in range(20)]

if not os.path.exists(filename):
    df = pd.DataFrame(columns=columns)
    df.to_csv(filename, index=False)


landmarks_per_cycle = 21
total_cycles = len(col3) // landmarks_per_cycle

for c in range(total_cycles):

    start = c * landmarks_per_cycle
    end = start + landmarks_per_cycle

    differences_x = []
    differences_y = []

    for i in range(start, end - 1):
        diff_x = int(col3[i]) - int(col3[i+1])
        diff_y = int(col4[i]) - int(col4[i+1])
        differences_x.append(diff_x)
        differences_y.append(diff_y)

    cycle_number = c + 1

    row_x = [cycle_number, "x"] + differences_x
    row_y = [cycle_number, "y"] + differences_y

    df_append = pd.DataFrame([row_x, row_y], columns=columns)
    df_append.to_csv(filename, mode="a", header=False, index=False)
```
